## 1.환경준비

### (1) 라이브러리 로딩

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import *
from sklearn.preprocessing import MinMaxScaler

from keras.models import Sequential
from keras.layers import Dense
from keras.backend import clear_session
from keras.optimizers import Adam
```

* 함수 만들기

```
# 학습곡선 함수
def dl_history_plot(history):
    plt.figure(figsize=(10,6))
    plt.plot(history['loss'], label='train_err', marker = '.')
    plt.plot(history['val_loss'], label='val_err', marker = '.')

    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid()
    plt.show()
```

## 2.데이터 전처리

### (1) 데이터 정리

```
target = 'Attrition'
```

```
# 불필요한 변수 제거
data.drop('EmployeeNumber', axis = 1, inplace = True)
```

```
x = data.drop(target, axis = 1)
y = data.loc[:, target]
```

### (2) 가변수화
* 범주형 데이터이면서 값이 0,1 로 되어 있는 것이 아니라면, 가변수화를 수행해야 합니다.
* 대상이 되는 변수에 대해서 가변수화를 수행해주세요.

```
dum_cols = ['BusinessTravel','Department','Education','EducationField','EnvironmentSatisfaction','Gender',
            'JobRole', 'JobInvolvement', 'JobSatisfaction', 'MaritalStatus', 'OverTime', 'RelationshipSatisfaction',
            'StockOptionLevel','WorkLifeBalance' ]

x = pd.get_dummies(x, columns = dum_cols ,drop_first = True)
x.head()
```

### (3) 데이터 분할
* train_test_split :
    * test_size : 0.# - 비율로 분할, 1보다 큰 자연수 - 갯수로 분할
    * train_size로 지정도 가능.

```
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size = 200, random_state = 2022)
```

### (4) 스케일링

```
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_val = scaler.transform(x_val)
```

## 4.모델링

### (1) 모델1
* 다양한 구조의 모델 2개 이상을 설계하시오. (히든레이어, 노드 수 조절)

```
nfeatures = x_train.shape[1]
nfeatures
```

```
clear_session()

# base model
b_model = Sequential(Dense(1, input_shape = (nfeatures,), activation = 'sigmoid'))

# compile
b_model.compile(optimizer = Adam(learning_rate = 0.01), loss = 'binary_crossentropy')
```

```
# 학습(, verbose = 0) <-- 에포크마다 성능 로그 화면에 출력하는것 비활성화
hist = b_model.fit(x_train, y_train, epochs = 50, validation_split = .2
                    , verbose = 0).history

# 학습 곡선
dl_history_plot(hist)
```

```
# 예측 및 평가
pred0 = b_model.predict(x_val)
pred0 = np.where(pred0 >= .5, 1, 0)
print(confusion_matrix(y_val, pred0))
print(classification_report(y_val, pred0))
```

### (2) 모델2

* 레이어를 늘려가면서 시험하기

```
clear_session()

model = Sequential([Dense(8, input_shape = (nfeatures,), activation = 'relu'),
                    Dense(1, activation = 'sigmoid')])

# 위 코드와 동일 : model.add로 레이어를 하나씩 추가할 수 있음.
model = Sequential()
model.add(Dense(8, input_shape = (nfeatures,), activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))

# compile
model.compile(optimizer = Adam(learning_rate = 0.01), loss = 'binary_crossentropy')
```

```
# 히든레이어 n개 짜리 모델 생성
n = 10
clear_session()

model = Sequential()
model.add(Dense(8, input_shape = (nfeatures,), activation = 'relu'))

for i in range(n-1):
    model.add(Dense(8, activation = 'relu'))

model.add(Dense(1, activation = 'sigmoid'))
model.summary()
```

```
def tune_layer(n) :
    clear_session()

    model = Sequential()
    model.add(Dense(8, input_shape = (nfeatures,), activation = 'relu'))

    for i in range(n-1):
        model.add(Dense(8, activation = 'relu'))

    model.add(Dense(1, activation = 'sigmoid'))

    model.compile(optimizer = Adam(learning_rate = 0.01), loss = 'binary_crossentropy')

    model.fit(x_train, y_train, epochs = 50, validation_split = .2, verbose = 0)

    pred = model.predict(x_val)
    pred = np.where(pred >= .5, 1, 0)
    acc = accuracy_score(y_val, pred)

    return acc
```
